### Dropping Unncessary things and syntactic changes

```python
df = df.drop(columns=['About', 'Number of Passengers ', 'Previous names'])
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

### Better Visualizing Data

```python
df['distance_km'] = df['distance_from_previous_of_the_line'].str.replace(' KM', '').str.strip()
df['travel_time_min'] = df['time_taken_from_previous_of_the_line'].str.replace(' min', '').str.strip()
```

```python
df = df.drop(columns=['distance_from_previous_of_the_line', 'time_taken_from_previous_of_the_line'])
```

### Dtype Conversion

```python
df['distance_km'] = pd.to_numeric(df['distance_km'], errors='coerce')
df['travel_time_min'] = pd.to_numeric(df['travel_time_min'], errors='coerce')
```

## The project is based on 3 Main lines ie Central,harbour,western not the outskirts
