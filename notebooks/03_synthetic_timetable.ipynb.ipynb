## Target 1: Western Line

```python
western = df[df['line'] == 'Western'].copy()
western.shape
```

```python
western[['station', 'travel_time_min']]
```

## Western Full timetable

```python
from datetime import datetime, timedelta

trains = []
start_time = datetime.strptime("04:00", "%H:%M")
end_time = datetime.strptime("23:00", "%H:%M")
frequency = 10  # every 10 minutes

train_no = 1
current_start = start_time

while current_start <= end_time:
    current_time = current_start
    for i, row in western.iterrows():
        current_time += timedelta(minutes=row['travel_time_min'])
        trains.append({
            'train_id': f'WR_{str(train_no).zfill(3)}',
            'station': row['station'],
            'station_code': row['station_code'],
            'line': 'Western',
            'arrival_time': current_time.strftime("%H:%M"),
            'stop_number': list(western.index).index(i) + 1
        })
    train_no += 1
    current_start += timedelta(minutes=frequency)

timetable_wr = pd.DataFrame(trains)
print(timetable_wr.shape)
```
